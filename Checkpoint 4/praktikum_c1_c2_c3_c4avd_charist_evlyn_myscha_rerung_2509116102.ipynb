### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Data Construction**

Pada tahap data construction ini saya menambahkan satu kolom baru yaitu Sleep_Quality. Kolom ini berguna untuk membantu kita memahami apakah durasi tidur seseorang berhubungan dengan kondisi kesehatan mentalnya. Dengan adanya kolom ini, analisis menjadi lebih mudah karena kita tidak hanya melihat angka jam tidur, tetapi juga kualitas tidur yang lebih manusiawi dan relevan dengan kehidupan sehari-hari.


```python
def categorize_sleep_quality(sleep_hours):
    if sleep_hours > 8:
        return 'Baik'
    elif sleep_hours >= 6:
        return 'Normal'
    else:
        return 'Kurang'

df['Sleep_Quality'] = df['Sleep_Duration_Hours'].apply(categorize_sleep_quality)
```

### **BoxPlot**

Berikut ini ialah BoxPlot yang saya buat untuk melihat distribusi durasi penggunaan media sosial per hari pada pengguna. Grafik ini digunakan untuk melihat sebaran data, nilai tengah, serta kemungkinan adanya nilai yang lebih tinggi atau lebih rendah dari kebanyakan data. Dengan grafik ini, kita dapat memahami bagaimana pola penggunaan media sosial dalam dataset yang digunakan.

```python
top3platform = df['Primary_Platform'].value_counts().head(3).index
df_top3platform = df[df['Primary_Platform'].isin(top3platform)]

plt.figure(figsize=(12, 7))
df_top3platform.boxplot(column='Daily_Screen_Time_Hours', by='Primary_Platform',
                        ax=plt.gca(), grid=False, patch_artist=True,
                        medianprops=dict(color='lightcoral'))

plt.title('TOP 3 PLATFORM MEDIA SOSIAL BERDASARKAN DURASI PENGGUNAAN', fontsize=16)
plt.xlabel('PLATFORM MEDIA SOSIAL', fontsize=12)
plt.ylabel('DURASI PENGGUNAAN (JAM)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

### **Heatmap**

Berikut ini ialah Heatmap yang saya buat dan gunakan untuk melihat hubungan atau korelasi antara beberapa variabel numerik pada dataset, yaitu Daily_Screen_Time_Hours (durasi penggunaan media sosial per hari) dan Sleep_Duration_Hours (durasi tidur). Warna pada heatmap menunjukkan seberapa kuat hubungan antar variabel: semakin terang atau semakin kuat warnanya, maka hubungan antar variabel tersebut juga semakin kuat. Dari grafik dapat dilihat bagaimana kedua variabel tersebut saling berkaitan, sehingga membantu dalam memahami apakah durasi penggunaan media sosial memiliki hubungan tertentu dengan durasi tidur pengguna.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Daily_Screen_Time_Hours', 'Sleep_Duration_Hours']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')

plt.title('Korelasi Durasi Penggunaan Media Sosial dan Durasi Tidur')
plt.show()
```

Insight :

- Heatmap menunjukkan hubungan antara durasi penggunaan media sosial dan durasi tidur. Korelasi yang terlihat tidak terlalu kuat, sehingga kedua variabel tidak memiliki hubungan yang signifikan.

- Hal ini menunjukkan bahwa lamanya penggunaan media sosial tidak selalu menentukan berapa lama seseorang tidur.

Action :

- Pengguna media sosial tetap perlu memperhatikan durasi penggunaan media sosial agar tidak mengganggu waktu istirahat.

- Menjaga keseimbangan antara aktivitas online dan waktu tidur yang cukup dapat membantu menjaga kesehatan dan produktivitas.
