### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **Eksplorasi Data (EDA)**

## **Comparison(Perbandingan)**

```python
screen_time_platform = df.groupby('Primary_Platform')['Daily_Screen_Time_Hours'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(
    x=screen_time_platform.index,
    y=screen_time_platform.values,
    palette='viridis'
)
plt.title('Rata-rata Screen Time Berdasarkan Platform')
plt.xlabel('Platform')
plt.ylabel('Rata-rata Screen Time (Jam)')
plt.xticks(rotation=45)
plt.show()
```

**Kesimpulan**

Berdasarkan grafik dapat dilihat bahwa rata-rata screen time berbeda di setiap platform. Platform seperti TikTok memiliki rata-rata screen time yang lebih tinggi dibandingkan platform lain, sementara Facebook berada pada tingkat yang lebih rendah. Hal ini menunjukkan adanya variasi intensitas penggunaan media sosial antar platform.


## **Composition/Komposisi**

```python
screen_time_platform = df.groupby('Primary_Platform')['Daily_Screen_Time_Hours'].sum().sort_values(ascending=False)
screen_time_top3 = screen_time_platform.head(3)

plt.figure(figsize=(10,8))
screen_time_top3.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Proporsi 3 Platform Teratas Berdasarkan Total Screen Time')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

**Kesimpulan**

Berdasarkan pie chart dapat dilihat bahwa proporsi total screen time tidak merata di setiap platform. Tiga platform teratas menyumbang porsi terbesar dari keseluruhan waktu penggunaan, menunjukkan bahwa sebagian besar waktu pengguna lebih banyak dihabiskan pada platform tersebut. Sementara itu, platform lain memiliki kontribusi yang lebih kecil terhadap total screen time.

## **Relationship/Hubungan**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    data=df[['Daily_Screen_Time_Hours', 'Sleep_Duration_Hours']].corr(),
    annot=True,
    cmap='viridis',
    fmt='.2f'
)
plt.title('Korelasi antara Screen Time dan Durasi Tidur')
plt.show()
```

**Kesimpulan**\
Berdasarkan heatmap korelasi antara Daily Screen Time dan Sleep Duration Hours, terlihat adanya hubungan yang cenderung negatif. Artinya, semakin lama seseorang menghabiskan waktu di layar setiap harinya, biasanya durasi tidurnya semakin pendek.
ini menunjukkan bahwa screen time berlebihan dapat berdampak pada pola tidur pengguna. Dengan kata lain, berdasarkan visualisasi tersebut, ada kecenderungan bahwa penggunaan media sosial yang tinggi berkaitan dengan berkurangnya waktu tidur.
